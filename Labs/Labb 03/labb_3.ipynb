```python
import matplotlib.pyplot as plt
import numpy as np

unlabelled_data = "./unlabelled_data.csv"  

x_values = []
y_values = []


with open(unlabelled_data, "r") as f_read:
    for line in f_read:
        x, y = map(float, line.split(","))
        x_values.append(x)
        y_values.append(y)
        plt.scatter(x, y, color='black')


# After looking over the scatterplot I figured that k = 1 and m = 0. 
# Entering these values into the algorithm "y = k(x) + m" gives us y = 1x + 0, which simplified is y = x. 
# Since y = x I can use the x_values when finding the highest and lowest values and plot a line based on those values
plt.plot([min(x_values), max(x_values)], [min(x_values), max(x_values)], color='red', label='y = x')



plt.show()
```
